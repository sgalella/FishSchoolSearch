# FishSchoolSearch

For more information of the algorithm, check [_A novel algorithm based on fish school behavior_](https://ieeexplore.ieee.org/document/4811695).

```python
import sys
sys.path.append('..')

import numpy as np
import fish_school_search as fs
```

```python
resolution = 100
limits = [-5, 5, -3, 3]  # x_min, x_max, y_min, y_max
num_iterations = 50
landscape = fs.SphereLandscape(limits, resolution)
```

## Show last search state

```python
np.random.seed(1234)
search = fs.FishSchoolSearch(landscape, num_individuals=10)
fs.VisualizeSearch.show_last(search, num_iterations)
```

## Show iterative search

```python
%matplotlib
np.random.seed(1234)
search = fs.FishSchoolSearch(landscape, num_individuals=10)
fs.VisualizeSearch.show_all(search, num_iterations)
```

## Plot best fitness achieved¶

```python
np.random.seed(1234)
search = fs.FishSchoolSearch(landscape, num_individuals=10)
for _ in range(num_iterations):
    search.update(num_iterations)
print(search.best_fitness)
```
